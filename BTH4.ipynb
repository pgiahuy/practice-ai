```python
data = {
    "MSSV": [2301, 2302, 2401],
    "HoTen": ["Ly Thi Hong", "Nguyen Van Hai", "Le Phuong Anh"],
    "NamSinh": [2000, 2002, 2001],
    "DiaChi": ["HCM", "VT", "DN"]
}

d = pd.DataFrame(data)
print(d)
print()
d['GioiTinh'] = ['','','']
s1 = {
    'MSSV' : 2306,
    'HoTen': 'Nguyen Van Tai', 
    'NamSinh': 2002,
    'DiaChi':'NT',
    'GioiTinh':'Nam'
}
s2 = {
    'MSSV' : 2404,
    'HoTen': 'Le Van Nam', 
    'NamSinh': 2001,
    'DiaChi':'VT',
    'GioiTinh':'Nam'
}
s3 = {
    'MSSV' : 2304,
    'HoTen': 'Phan Hong Tham', 
    'NamSinh': 2001,
    'DiaChi':'HCM',
    'GioiTinh':'Nu'
}
d_new = pd.DataFrame([s1, s2, s3])

print()
print(d)
print()
print("---new---")
print(d_new)


print(d)
print()
print("---full---")
print("----------")
df = pd.concat([d, d_new], ignore_index=True)
print(df)

d_nam = df[df['GioiTinh'].isin(['Nam'])]
print("---Nam---")
print(d_nam)

print()
print("---Nu---")
d_nu = df[(df['GioiTinh'] == 'Nu') & (df['DiaChi'] == 'HCM')]
print(d_nu)
print("So luong sv nu o HMC:",len(d_nu))

d_nu.to_csv('sv_nu_HCM.txt', encoding='utf-8', index = False, header ='True')
```
