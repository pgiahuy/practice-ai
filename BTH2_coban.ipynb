```python
# 1. Viết chương trình tính tổng 5 số được nhập vào từ bàn phím.
tong = 0
for i in range(1,6):
    tong += int(input("Nhập số thứ " + str(i) + " : "))

print("Sum =",tong )
```

```python
# 2. Viết chương trình tính tổng n số được nhập vào từ bàn phím, kết thúc bằng 
# phím ‘e’ hoặc ‘E’.
tong =0
while True:
    s = input("Nhập số : ")
    if s.lower()== "e":
        break
    try:
        tong += int(s)
    except ValueError:
        print("nhập số hợp lệ!")
print("Sum =",tong )
    
```

```python
# 3. Viết chương trình tính tổng các số chẳn dương được nhập vào từ bàn phím.
# Khi chạy chương trình: xuất ra các số vừa nhập vào, và kết quả.
tong =0
arr = []
while (inp := input("Nhập số : ")).lower() != "e" :
    # inp = input("Nhập số : ")
    # if inp.lower()== "e":
    #     break
    try:
        s= int(inp)
        if s > 0 and s %2==0:
            tong += s
            arr.append(s)
    except ValueError:
        print("nhập số hợp lệ!")
        
print("Các số chẵn dương:", *arr)
print("Sum =",tong )
```

```python
# 4. Viết chương trình tính tổng các số từ 1 đến n. n được nhập vào từ bàn phím.
n = int(input("Nhập số lượng số: "))
tong =0
for i in range (1,n+1):
    inp = input("Nhập số thứ " +str(i)+": ")
    try:
        s= int(inp)
        tong += s
    except ValueError:
        print("nhập số hợp lệ!")
print("Sum =",tong )
        
    
```

```python
# 9. Viết chương trình đếm số chữ tìm được trong một đoạn văn.
# Ví dụ: Tìm chữ 'Python' có trong đoạn văn nào đó, và hiển thị số lần xuất hiện của chữ Python.
cau = input("Nhập câu: ").lower()
kw = input("Nhập từ cần tìm: ").lower()

for ky_tu in ",.!?;:'\"":
    cau = cau.replace(ky_tu, " ")
    
ds_tu = cau.split()
so_lan_xh = ds_tu.count(kw)

print(*ds_tu)

if so_lan_xh == 0:
    print("Không tìm thấy !")
else:
    print(kw + " xuất hiện " + str(so_lan_xh) + " lần")

```

```python
# 10.Viết chương trình thay thế chuỗi trong một đoạn văn (không sử dụng hàm replace có sẵn)
# Ví dụ: Tìm tất cả các chữ ‘data’ có trong đoạn văn và thay thế bằng chữ ‘dữ liệu’. Sau đó xuất ra chuỗi mới sau khi đã thay thế.
cau = input("Nhập câu: ").lower()
old_word = input("Nhập từ cần thay: ").lower()
new_word = input("Nhập từ mới: ")

for ky_tu in ",.!?;:'\"":
    cau = cau.replace(ky_tu, " ")

ds_tu = cau.split()

for i in range(len(ds_tu)):
    if ds_tu[i] == old_word:
        ds_tu[i] = new_word
        
print(*ds_tu)
```
