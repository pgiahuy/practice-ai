```python
v = np.array([87, 46, 87, 62, 11, 32])
print(v)
print("Max =", end=" ")
print(np.max(v))
print("Min =", end=" ")
print(np.min(v))
v = np.sort(v)
print(v)
print(v[0::2])
print(v[3],end=" ")
print("la so", end=" ")
if v[3] %2==0:
    print("Chan")
else:
    print("le")

c = np.sum(v%2==0)
print("so luong so chan tron mang =",c)
print("Tong cac phan tu =", np.sum(v))
print("mean (trung binh) =", round(np.mean(v),2))
print("median (trung vi) =", np.median(v))


mean_v = np.mean(v)
print(v)
v1 = v[v!=62]
print(v1)

print(mean_v)
v1[v1==87] = mean_v
print(v1)
```
