```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('student.csv', header=None, sep=";", skiprows=1)
df.columns = ['ID', 'Name', 'BOD', 'Gender', 'Mark', 'Class']
df = pd.DataFrame(df)
df.replace(['M','F'],['Nam','Nữ'], inplace=True)



grouped = df.groupby(['Class', 'Gender']).size().unstack(fill_value=0)
print(grouped)
classes = grouped.index.tolist()
print("Class", classes)


X_axis = np.arange(len(classes))
nu = grouped['Nữ'].tolist()
nam = grouped['Nam'].tolist()



plt.bar(X_axis - 0.2, nu, width=0.4, label='Nữ')
plt.bar(X_axis + 0.2, nam, width=0.4, label='Nam')

plt.xticks(X_axis, classes, rotation=30)
plt.xlabel("Lớp")
plt.ylabel("Số lượng sinh viên")
plt.title("Số sinh viên Nam/Nữ theo từng lớp")

plt.legend()
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('student.csv', header=None, sep=";", skiprows=1)
df.columns = ['ID', 'Name', 'BOD', 'Gender', 'Mark', 'Class']


df['Gender'] = df['Gender'].replace({'M':'Nam', 'F':'Nữ'})


df['Mark'] = pd.to_numeric(df['Mark'], errors='coerce')

sns.boxplot(x='Class', y='Mark', hue='Gender', data=df)

plt.xticks(rotation=40)
plt.show()
```
