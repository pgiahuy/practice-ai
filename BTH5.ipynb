```python
df = pd.read_csv('student.csv', header=None, sep=";", skiprows=1)
df.columns = ['ID', 'Name', 'BOD', 'Gender', 'Mark', 'Class']
df = pd.DataFrame(df)
df.replace(['F','M'],['Nam','Nữ'], inplace=True)

df['Mark'] = pd.to_numeric(df['Mark'], errors='coerce')


diem_trung_binh = round(np.mean(df['Mark']),1)
# df.head(20)
# df['Mark'].replace(np.nan, diem_trung_binh, inplace=True)

df['Mark'] = df['Mark'].fillna(diem_trung_binh)

df = df.rename(columns={'ID': 'MSSV', 'Name':'Ho_Ten','BOD': 'Ngay_Sinh','Gender':'Gioi_Tinh','Mark': 'Diem','Class': 'Lop',})

df.head(5)

sv_nam = df
sv_nam = sv_nam[sv_nam['Gioi_Tinh'] == 'Nam']
sv_nam = sv_nam.drop(['Ngay_Sinh','Gioi_Tinh','Diem'], axis=1)
sv_nam.tail(50)

sv_nam.to_csv('bth5_sv.csv', encoding='utf-8', index=False, header=True)

new = pd.read_csv('bth5_sv.csv', header=None, skiprows=1)
print(new)
```
