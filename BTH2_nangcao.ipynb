```python
text = "jabbawocky"
na = ["a","o", "e","u","i"]
count_1 = 0
count_2 = 0
for i in text:
    if i in na:
        continue
    count_2+=1

for i in text:
    if i not in na:
        count_1+=1

print("Đếm cách 1:", count_1)
print("Đếm cách 2:", count_2)
```
